```python
import numpy as np
import seaborn as sns; sns.set(color_codes=True)
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(15,10)})
```

```python
acc = np.load("plot/acc.npy")
val_acc = np.load("plot/val_acc.npy")
loss = np.load("plot/loss.npy")
val_loss = np.load("plot/val_loss.npy")
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})

ax = sns.tsplot(data=val_loss, ci="sd", color="g", legend=True, condition="Test")
ax = sns.tsplot(data=loss, ci="sd", color="r", legend=True, condition="Train")
ax.set(xlabel='Epochs', ylabel='Loss')

plt.savefig("model_loss.png")
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})

ax = sns.tsplot(data=val_acc, ci="sd", color='g', legend=True, condition="Test")
ax = sns.tsplot(data=acc, ci="sd", color='r', legend=True, condition="Train")
ax.set(xlabel='Epochs', ylabel='Accuracy')

plt.savefig("model_acc.png")
```
